## Set API keys via Colab Secrets

```python
from google.colab import userdata

# set LLM Api Key
openrouter_api_key = userdata.get('OPENROUTER_API_KEY')
```

## Send HTTP request to OpenRouter API and print result

```python
import requests
import json

def send_request(prompt: str):
  response = requests.post(
    url="https://openrouter.ai/api/v1/chat/completions",
    headers={
      "Authorization": f"Bearer {openrouter_api_key}"
    },
    data=json.dumps({
      "model": "openrouter/auto",
      "messages": [
        {
          "role": "user",
          "content": prompt
        }
      ]
    })
  )


  try:
    response_json = response.json()

    print(f"Provider: {response_json.get('provider')}")
    print(f"Model: {response_json.get('model')}")

    if 'choices' in response_json and isinstance(response_json['choices'], list):
      for choice in response_json['choices']:
          if 'message' in choice and isinstance(choice['message'], dict):
            print(f"  Message Role: {choice['message'].get('role')}")
            print(f"  Message Content: {choice['message'].get('content')}")
          else:
            print("  Message data not found or in unexpected format.")
    else:
        print("'choices' key not found in the response or not a list.")

  except json.JSONDecodeError:
    print("Error: Invalid JSON response received.")
  except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
send_request("What is the meaning of life, the universe and everything?")
```
